```python
import random
import numpy as np

random.seed(1234)
np.random.seed((1234))
```

```python
import scipy.stats

from numpy.random import randn

import matplotlib.pyplot as plt
```

```python
num_data = 200
e_z = randn(num_data)
e_y = randn(num_data)

Z = e_z
Y = 2 * Z + e_y
```

```python
np.corrcoef(Z,Y)
```

```python
X_std=scipy.stats.zscore(Z)
Z_std=scipy.stats.zscore(Y)

plt.scatter(Z_std,X_std)
```

```python
num_data = 200
e_x = randn(num_data)
e_y = randn(num_data)
e_z = randn(num_data)

Z = 3.3 * e_x + e_z
Y = 3.3 * e_x + e_y

np.corrcoef(Z,Y)
```

```python
X_std=scipy.stats.zscore(Z)
Z_std=scipy.stats.zscore(Y)

plt.scatter(Z_std,X_std)
```
